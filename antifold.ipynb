```python
import pandas as pd
import antifold
import antifold.main
from matplotlib import pyplot as plt
import seaborn as sns
```

```python

# Load model
model = antifold.main.load_model() # AntiFold
# model = antifold.main.load_model("ESM-IF1") # ESM-IF1 mode

# PDB directory
pdb_dir = "." # Current directory

# Assumes first chain heavy, second chain light
pdbs_csv = pd.DataFrame({
    "pdb": "5nb5",
    "Hchain": "H",
    "Lchain": "L",
}, index=[0])

# Sample from PDBs
df_logits_list = antifold.main.get_pdbs_logits(
    model=model,
    pdbs_csv_or_dataframe=pdbs_csv,
    pdb_dir=pdb_dir,
)

# Output log probabilites
df_logits_list[0]
```

```python
df_logits_listt = pd.DataFrame(df_logits_list[0])
dff_logits_listt=pd.DataFrame(df_logits_listt.iloc[:,5:])
```

```python
dff_logits_listt_cor=dff_logits_listt.corr()
dff_logits_listt_cor.style.background_gradient(cmap="coolwarm").format(precision=2)
```
